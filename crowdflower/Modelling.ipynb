## SVD - SVM

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn import decomposition, pipeline, metrics
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
# https://www.kaggle.com/abhishek/beating-the-benchmark
scl = StandardScaler()
for component in [460]:
  for C in [11]:

    svd = TruncatedSVD(n_components=component)
    svd.fit(X_train)
    X_train_svd = svd.transform(X_train)
    X_cv_svd = svd.transform(X_cv)
    scl.fit(X_train_svd)
    X_train_sc = scl.transform(X_train_svd)
    X_cv_sc = scl.transform(X_cv_svd)
    svm_model  = SVC(C = C)
    svm_model.fit(X_train_sc,y_train)
    train_pred = svm_model.predict(X_train_sc)
    cv_pred = svm_model.predict(X_cv_sc)
    print(component,C,)
    print(cohen_kappa_score(y_train,train_pred,weights = 'quadratic'))
    print(cohen_kappa_score(y_cv,cv_pred,weights = 'quadratic'))
    print('*'*50)



```

```python
svd = TruncatedSVD()

# Initialize the standard scaler 
scl = StandardScaler()

# We will use SVM here..
svm_model = SVC()

# Create the pipeline 
clf = pipeline.Pipeline([('svd', svd),
              ('scl', scl),
                        ('svm', svm_model)])

# Create a parameter grid to search for best parameters for everything in the pipeline
param_grid = {'svd__n_components' : [200, 400],
              'svm__C': [10, 12]}

# Kappa Scorer 
kappa_scorer = metrics.make_scorer(cohen_kappa_score,weights = 'quadratic', greater_is_better = True)

# Initialize Grid Search Model
model = GridSearchCV(estimator = clf, param_grid=param_grid, scoring=kappa_scorer,
                                  verbose=10, n_jobs=-1, iid=True, refit=True, cv=2)
                                  
# Fit Grid Search Model
model.fit(X_train, y_train)
print("Best score: %0.3f" % model.best_score_)
print("Best parameters set:")
best_parameters = model.best_estimator_.get_params()
for param_name in sorted(param_grid.keys()):
  print("\t%s: %r" % (param_name, best_parameters[param_name]))
```

```python
pred = model.predict(X_cv) 
# pred1 = np.round(pred)
print(cohen_kappa_score(y_cv,pred,weights = 'quadratic'))
pred = model.predict(X_test) 
# pred1 = np.round(pred)
print(cohen_kappa_score(y_test,pred,weights = 'quadratic'))
```
